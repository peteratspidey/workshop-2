## series
> series is a 1D indexed array capable of holding any data type .
### creating an series
there are diff ways to create a series


```python
my_series = pd.Series([1,34,65,78,91])
my_series
```

```python
type(my_series)
```

```python
my_series = pd.Series([1,34,65,78,91], index= ['a','b', 'c', 'd', 'e'])
my_series
```

```python
my_series['b']
```

```python
my_series.mean()
```

```python
my_series.median()
```

```python
my_series[my_series>my_series.mean()]
```
